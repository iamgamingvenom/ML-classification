```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('diabetes2.csv')
df.head(5)
```

```python
df.info()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
print(np.concatenate((lr_pred.reshape(len(lr_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
lr_cm = confusion_matrix(lr_pred, y_test)
lr_as = accuracy_score(lr_pred, y_test)
print('Linear Regression Model: ', lr_cm)
print('Linear Regression Model: ', lr_as)
```

```python
svc = SVC(kernel='linear', random_state=42)
svc.fit(x_train, y_train)
```

```python
svc_pred = svc.predict(x_test)
print(np.concatenate((svc_pred.reshape(len(svc_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
svc_cm = confusion_matrix(svc_pred, y_test)
svc_as = accuracy_score(svc_pred, y_test)
print('SVC model: ', svc_cm)
print('SVC model: ', svc_as)
```

```python
knn = KNeighborsClassifier(n_neighbors=4, metric='minkowski', p=2)
knn.fit(x_train, y_train)
```

```python
knn_pred = knn.predict(x_test)
print(np.concatenate((knn_pred.reshape(len(knn_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
knn_cm = confusion_matrix(knn_pred, y_test)
knn_as = accuracy_score(knn_pred, y_test)
print('KNN model', knn_cm)
print('KNN model', knn_as)
```
