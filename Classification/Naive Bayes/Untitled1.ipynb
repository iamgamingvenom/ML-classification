```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head(5)
```

```python
df.info()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

#### Splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

#### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

#### Dimensionality reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=42)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test)
```

#### Kfold cross validation

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(np.concatenate((y_test.reshape(len(y_test), 1), y_pred.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred, y_test)
print(cm)
accuracy_score(y_pred, y_test)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = model, X = x_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from xgboost import XGBClassifier
boosting_model = GaussianNB()
boosting_model.fit(x_train, y_train)

y_pred_boost = boosting_model.predict(x_test)
print(f"Boosting Test Accuracy: {accuracy_score(y_test, y_pred_boost):.4f}")
```
